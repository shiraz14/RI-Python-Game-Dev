## 📌 Section 5 — Loops: Repeating Actions

**`while` loop** — repeats while a condition is `True`:
```python
count = 1
while count <= 3:
    print("Count:", count)
    count = count + 1
```

**`for` loop** — repeats for each item in a sequence:
```python
for i in range(5):
    print(i)   # prints 0, 1, 2, 3, 4
```

> ⚠️ Always update the variable in a `while` loop — forgetting to do so creates an **infinite loop**!

## 📌 Section 6 — Lists: Storing Multiple Items

A **list** holds multiple values in order, inside square brackets. Indexing starts at **0**.

```python
enemies = ["Goblin", "Dragon", "Slime"]
print(enemies[0])    # Goblin  (first item)
print(enemies[2])    # Slime   (third item)

enemies.append("Boss")   # add to end
enemies.remove("Slime")  # remove by value
print(len(enemies))      # number of items → 3
```

## 📌 Section 7 — Functions: Reusing Code

A **function** is a named block of code you can call whenever you need it.

```python
def greet_player(name):
    print("Welcome,", name)

greet_player("Alice")   # Welcome, Alice
greet_player("Bob")     # Welcome, Bob
```

Functions can also **return** a value:
```python
def double(number):
    return number * 2

result = double(7)
print(result)   # 14
```

## 📌 Section 9 — Comments: Making Code Clear

Comments start with `#` and are **ignored by Python** — they explain your code to humans.

```python
# This variable tracks how many lives the player has
lives = 3

lives = lives - 1   # subtract one life when player dies

# Check if the game should end
if lives == 0:
    print("Game Over")
```

> 💡 Good comments explain **WHY** code does something, not just **WHAT** it does.
